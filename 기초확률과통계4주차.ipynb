```python
import numpy as np
from statistics import mode
from scipy import stats
examscore=[100,99,98,97,96,55,66,77,33,45]
meanscore=np.mean(examscore)
medianscore=np.median(examscore)
try:
    modescore = mode(examscore)
except:
    modescore = stats.mode(examscore)[0][0]
print(f"평균 점수: {meanscore}")
print(f"중앙값 점수: {medianscore}")
print(f"최빈값 점수: {modescore}")
```

```python
import numpy as np
from scipy.stats import variation
data = [1,2,3,4,5,3,4,5,3,4,5]
data_range = np.max(data) - np.min(data)
variance = np.var(data, ddof=0)
sample_variance = np.var(data, ddof=1)
std_deviation = np.std(data, ddof=0)
sample_std_deviation = np.std(data, ddof=1)
cv = variation(data)
print(f"범위: {data_range}")
print(f"분산(모집단): {variance}")
print(f"분산(표본): {sample_variance}")
print(f"표준편차(모집단): {std_deviation}")
print(f"표준편차(표본): {sample_std_deviation}")
print(f"변동계수: {cv}")
```
