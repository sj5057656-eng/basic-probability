```python
def conditional_probability(prob_a_and_b, prob_b):
 """
조건부확률을계산하는함수
:param prob_a_and_b: 사건A와B가동시에일어날확
률
:param prob_b: 사건B가일어날확률
:return: 조건부확률P(A|B)
 """
 return prob_a_and_b / prob_b

prob_a = 0.6
prob_b = 0.4
prob_a_and_b = 0.3
prob_a_given_b = conditional_probability(prob_a_and_b, prob_b)
print("우유를산경우에빵을살확률:", prob_a_given_b)  #"""을 붙여야 밑의 :이 인식됨
```
